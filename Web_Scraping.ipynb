```python
import requests
from bs4 import BeautifulSoup
```

```python
r1 = requests.get(input())
coverpage = r1.content
```

```python
(soup1 := BeautifulSoup(coverpage, 'html5lib'))
```

```python
(Ftitle := soup1.find_all('h1'))
```

```python
(Ftext := soup1.find_all('p'))
```

```python
title=Ftitle[0].get_text()
text=''
for i in Ftext: 
    text+=i.get_text()
```

```python
print(title,text,sep='\n\n\n')
```
